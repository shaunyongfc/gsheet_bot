```python
from urllib.request import urlopen
import json

response = urlopen("https://raw.githubusercontent.com/shalzuth/wotv-ffbe-dump/master/jpdata/Skill.json")
json_data = response.read().decode('utf-8', 'replace')

d = json.loads(json_data)
```
